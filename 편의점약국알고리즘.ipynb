```python
import numpy as np
import numpy.random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 10
conv = 100
mat_a = np.array([[np.random.randint(180000, 216000, size= n)],
                  [np.random.randint(435000, 466000, size= n)]])
```

```python
data_c = pd.read_csv('C:\\Users\\Administrator\\Desktop\\5조데이터\\conveninece_store_drug_store_total.csv', encoding='euc-kr')
data_c2 = data_c.filter(items=['x', 'y'])
```

```python
rl = mat_a.transpose(1, 2, 0)
rl_2 = rl.reshape(1 * n, 2)
```

```python
list_dist = np.array([])
for i in range(0, n):
    for j in range(1, conv):
            distance = np.linalg.norm(rl_2[i] - data_c2[j-1:j], ord = 2)
            list_dist = np.append(list_dist, distance)
            list_dist
```

```python
np.savetxt("distance_3.csv", list_dist, fmt= '%d', delimiter=',')
```

```python
splited = np.split(list_dist, n)
upflow = np.sort(splited, axis=1)
sliced = np.sort(upflow, axis=1)
min_5 = sliced[:, :5]
```
